```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_csv("complete_dataset.csv")
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["demand"], y=df["RRP"], alpha=0.5)

X = df["demand"].values.reshape(-1,1)
y = df["RRP"].values
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

plt.plot(df["demand"], y_pred, color='red', label="Trend Line")
plt.xlabel("Electricity Demand (MW)")
plt.ylabel("Electricity Price (AUD)")
plt.title("Electricity Demand vs. Price with Trend Line")
plt.legend()
plt.show()
```

## <font color='green'> Anomaly Detection & Outlier Analysis in Electricity Demand </font>

```python
import pandas as pd

df = pd.read_csv("complete_dataset.csv")
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.dropna(inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')
df_daily = df.resample('D').mean()
print(df_daily.head())
```

## <font color='green'>Outlier Detection</font>

```python
Q1 = df_daily["demand"].quantile(0.25)
Q3 = df_daily["demand"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df_daily[(df_daily["demand"] < lower_bound) | (df_daily["demand"] > upper_bound)]
print("Detected Outliers:")
print(outliers)
```

#### High outliers: Sudden demand spikes due to heatwaves, system failures, or events. Low outliers: Sharp demand drops due to holidays, lockdowns, or technical issues.



### Visualize Outliers

```python
plt.figure(figsize=(12,6))
plt.plot(df_daily.index, df_daily["demand"], label="Electricity Demand", color="black")
plt.scatter(outliers.index, outliers["demand"], color="red", label="Outliers", zorder=3)
plt.xlabel("Date")
plt.ylabel("Electricity Demand (MW)")
plt.title("Electricity Demand with Detected Outliers")
plt.legend()
plt.grid(True)
plt.show()
```

     If anomalies match temperature spikes, they are likely due to heatwaves/cold spells.

### Using Median Replace Outlier

```python
import pandas as pd
df = pd.read_csv("complete_dataset.csv")  

Q1 = df["demand"].quantile(0.25) 
Q3 = df["demand"].quantile(0.75)  
IQR = Q3 - Q1 
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
median_value = df["demand"].median()
df["demand"] = df["demand"].apply(lambda x: median_value if (x < lower_bound or x > upper_bound) else x)
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.boxplot(df["demand"])
plt.title("Electricity Demand After Outlier Replace")
plt.show()
```

### <font color='red'>Demand and Price Outlier Replace By Median</font>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("complete_dataset.csv")  

def replace_outliers_with_median(column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find median value
    median_value = df[column].median()

    # Replace outliers with median
    df[column] = df[column].apply(lambda x: median_value if (x < lower_bound or x > upper_bound) else x)

# Apply function to both 'demand' and 'price' columns
replace_outliers_with_median("demand")
replace_outliers_with_median("RRP")

# Print cleaned data
print(df.head())

# Boxplot to check outliers after replacement
plt.figure(figsize=(10,5))
plt.boxplot([df["demand"], df["RRP"]], labels=["Demand", "Price"])
plt.title("Electricity Demand & Price After Outlier Replace")
plt.show()
```
