```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
df = pd.read_csv("cleaned_dataset.csv")

df['date'] = pd.to_datetime(df['date'])

df['z_score'] = np.abs(stats.zscore(df['demand']))

df['is_anomaly'] = df['z_score'] > 3

plt.figure(figsize=(14, 10))

plt.subplot(2, 1, 1)
sns.scatterplot(
    x='date', y='demand', data=df, hue='is_anomaly', 
    palette={False: 'teal', True: 'red'}, legend='full'
)
plt.title('Electricity Demand Anomalies Over Time')
plt.xlabel('Date')
plt.ylabel('Electricity Demand')

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['Normal', 'Anomaly'], title='Anomaly', loc='upper right')

plt.subplot(2, 1, 2)
sns.scatterplot(
    x='min_temperature', y='demand', data=df, hue='is_anomaly', 
    palette={False: 'teal', True: 'red'}, legend='full'
)
plt.title('Electricity Demand vs. Extreme Weather (Min Temperature)')
plt.xlabel('Min Temperature')
plt.ylabel('Electricity Demand')

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['Normal', 'Anomaly'], title='Anomaly', loc='upper right')

plt.tight_layout()
plt.show()
```

**1.Demand Anomalies Over Time:**

A.A scatter plot marks normal demand in tale and anomalies in red to highlight unusual spikes or drops.

**2.Demand vs. Extreme Weather (Min Temperature):**

A.This plot shows how minimum temperature correlates with demand, identifying whether extreme weather contributes to anomalies.

```python
import numpy as np

q1 = df['demand'].quantile(0.25)
q3 = df['demand'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers_before = df[(df['demand'] < lower_bound) | (df['demand'] > upper_bound)]

df['demand_capped'] = np.where(df['demand'] > upper_bound, upper_bound,
                                 np.where(df['demand'] < lower_bound, lower_bound, df['demand']))

outliers_after = df[(df['demand_capped'] < lower_bound) | (df['demand_capped'] > upper_bound)]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['demand'])
plt.title('Before Outlier Treatment')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['demand_capped'])
plt.title('After Outlier Treatment')

plt.show()

print(f"Number of outliers before capping: {len(outliers_before)}")
print(f"Number of outliers after capping: {len(outliers_after)}")
```

**As we can see that there are few outliers in the electricity demand. By using the IQR method we can handle the outliers. The left boxplot shows outliers before treatment, while the right boxplot confirms they are reoved after capping.**

**B).Scatter Plot of Demand vs. RRP to Highlight Negative Pricing Periods:**

1.Negative prices (red points) occur at different demand levels, meaning market factors influence pricing beyond just demand.

**C).Time Series of RRP with Highlighted Negative Price Periods:**

1.Excess Renewable Generation: High wind or solar energy during low demand periods can lead to oversupply.

2.Low Demand Periods: Nights, weekends, and mild weather conditions reduce electricity consumption, causing price drops. 

3.Grid Constraints: Some power plants cannot quickly adjust output, leading to excess supply even when demand is low.

```python
df['date'] = pd.to_datetime(df['date'])

df['month'] = df['date'].dt.month_name()
df['day_of_week'] = df['date'].dt.day_name()

monthly_demand = df.groupby('month')['demand'].mean()
monthly_demand = monthly_demand.reindex([
    'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 
    'September', 'October', 'November', 'December'
])

weekday_demand = df.groupby('day_of_week')['demand'].mean()
weekday_demand = weekday_demand.reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

fig, axes = plt.subplots(2, 1, figsize=(10, 10))

axes[0].bar(monthly_demand.index, monthly_demand.values, color='skyblue')
axes[0].set_title("Average Electricity Demand by Month")
axes[0].set_ylabel("Average Demand")
axes[0].tick_params(axis='x', rotation=45)

axes[1].bar(weekday_demand.index, weekday_demand.values, color='salmon')
axes[1].set_title("Average Electricity Demand by Day of the Week")
axes[1].set_ylabel("Average Demand")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**A).Seasonal Trend: Average Electricity Demand by Month:**

1.Electricity demand peaks in winter months (June-August), likely due to increased heating.

2.Lower demand in summer months (December-February) when temperatures are higher.

**B).Average Demand by Day of the Week:**

1.Weekdays (Monday-Friday) have higher demand due to business and school activities.

2.Weekends (Saturday-Sunday) show lower demand, likely reflecting residential energy usage patterns.
