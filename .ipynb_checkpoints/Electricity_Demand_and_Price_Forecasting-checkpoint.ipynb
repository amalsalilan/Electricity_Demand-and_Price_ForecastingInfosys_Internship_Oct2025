**Necessary Features:**

**Demand** - Demand, demand_pos, demand_neg

**price** - RRP = Regional Reference Price, RRP_pos, RRP_neg

**Weather** - min_temp, max_temp, solar_exposure, rainfall, school_day, holiday

#### Columns of Dataset:
**RRP:** Regional Reference Price in AUD.  
**demand_pos_RRP:** Electricity demand at positive RRP in megawatt-hours (MWh).  
**RRP_positive:** Positive Regional Reference Price in dollars.  
**demand_neg_RRP:** Electricity demand at negative RRP in megawatt-hours (MWh).  
**RRP_negative:** Negative Regional Reference Price in dollars.  
**frac_at_neg_RRP:** Fraction of time at negative RRP (unitless).  
**min_temperature:** Minimum daily temperature in degrees Celsius (°C).  
**max_temperature:** Maximum daily temperature in degrees Celsius (°C).  
**solar_exposure:** Daily solar exposure in megajoules per square meter (MJ/m²).  
**rainfall:** Daily rainfall in millimeters (mm).  
**school_day:** Indicates whether it is a school day (Yes/No).  
**holiday:** Indicates whether it is a holiday (Yes/No).

## Visualizing basic features and outliers

### Demand and Price Analysis

#### Explore daily demand variations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data['date'] = pd.to_datetime(data['date'])

# Explore daily demand variations
data['Day'] = data['date'].dt.day
daily_demand = data.groupby('Day')['demand'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=daily_demand.index, y=daily_demand.values)
plt.title('Average Daily Demand')
plt.xlabel('Day')
plt.ylabel('Average Demand')
plt.show()
```

#### Explore weekly demand variations

```python
# Explore weekly demand variations with colors
data['Weekday'] = data['date'].dt.weekday
weekly_demand = data.groupby('Weekday')['demand'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Weekday', y='demand', data=weekly_demand, palette='viridis', hue='Weekday', dodge=False)
plt.title('Average Weekly Demand')
plt.xlabel('Weekday')
plt.ylabel('Average Demand')
plt.legend([],[], frameon=False)
plt.show()
```

#### Explore monthly demand variations

```python
# Explore monthly demand variations
data['Month'] = data['date'].dt.month
monthly_demand = data.groupby('Month')['demand'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_demand.index, y=monthly_demand.values)
plt.title('Average Monthly Demand')
plt.xlabel('Month')
plt.ylabel('Average Demand')
plt.show()
```

#### Analyze the relationship between demand and electricity price (RRP)

```python
# Analyze the relationship between demand and electricity price (RRP)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='demand', y='RRP', data=data)
plt.title('Demand vs. Electricity Price (RRP)')
plt.xlabel('Demand')
plt.ylabel('Electricity Price (RRP)')
plt.show()

# Correlation analysis
correlation = data['demand'].corr(data['RRP'])
print(f'Correlation between Demand and Electricity Price (RRP): {correlation}')

# Identify periods with high demand and price spikes
high_demand_periods = data[(data['demand'] > data['demand'].quantile(0.75)) & (data['RRP'] > data['RRP'].quantile(0.75))]
print(high_demand_periods)
```

### Weather Impact Analysis

#### Correlate demand with temperature, solar exposure, and rainfall

```python
# Correlate demand with weather factors
weather_factors = ['min_temperature', 'max_temperature', 'solar_exposure', 'rainfall']
correlations = data[weather_factors + ['demand']].corr()['demand'].drop('demand')
print(correlations)

# Visualize correlations
plt.figure(figsize=(10, 6))
sns.heatmap(data[weather_factors + ['demand']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Demand and Weather Factors')
plt.show()
```

#### Analyze seasonal trends based on months in Australia

```python
# Create a new column for the season based on Australian months
def get_season(month):
    if month in [12, 1, 2]:
        return 'Summer'
    elif month in [3, 4, 5]:
        return 'Autumn'
    elif month in [6, 7, 8]:
        return 'Winter'
    else:
        return 'Spring'

data['Month'] = data['date'].dt.month
data['Season'] = data['Month'].apply(get_season)

# Analyze seasonal trends
seasonal_demand = data.groupby('Season')['demand'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='demand', data=seasonal_demand, palette='coolwarm', hue='Season', dodge=False)
plt.title('Average Demand by Season')
plt.xlabel('Season')
plt.ylabel('Average Demand')
plt.legend([],[], frameon=False)
plt.show()

print()

# Analyze demand under specific weather conditions
plt.figure(figsize=(14, 10))
for factor in weather_factors:
    plt.subplot(2, 2, weather_factors.index(factor) + 1)
    sns.scatterplot(x=factor, y='demand', data=data, alpha=0.5)
    plt.title(f'Demand vs {factor}')
    plt.xlabel(factor)
    plt.ylabel('Demand')
plt.tight_layout()
plt.show()
```

### Anomaly Detection and Outlier Analysis

#### Identify Demand Outliers and Investigate Potential Causes

```python
# Calculate Z-scores for demand
data['demand_zscore'] = stats.zscore(data['demand'])

# Identify outliers (Z-score > 3 or < -3)
data['is_anomaly'] = data['demand_zscore'].abs() > 3
outliers = data[data['is_anomaly']]

# Check if outliers coincide with holidays
external_events = outliers[outliers['holiday'] == 1]

# Check if outliers occur on non-school days
operational_issues = outliers[outliers['school_day'] == 0]

# Check if outliers occur due to extreme weather
extreme_weather = outliers[
    (outliers['min_temperature'] > data['min_temperature'].quantile(0.95)) |
    (outliers['max_temperature'] > data['max_temperature'].quantile(0.95))
]

# Print summary of potential causes
print("Potential Causes of Outliers:")
print(f"Holidays: {len(external_events)}")
print(f"Operational Issues (Non-School Days): {len(operational_issues)}")
print(f"Extreme Weather Events: {len(extreme_weather)}")
```

A correlation matrix shows the relationship between variables. The values range from -1 to +1

+1: Perfect positive correlation (variables move in the same direction).

0: No correlation (variables are independent).

-1: Perfect negative correlation (variables move in opposite directions).

**Handle Outliers:**

**Remove Outliers:** Delete the outlier data points from the dataset.

**Transform Data:** Apply transformations (e.g., log transformation) to reduce the impact of outliers.

**Cap and Floor:** Limit the range by setting upper and lower limits on the data.

**Impute Outliers:** Replace outliers with statistical measures like the mean or median.

**Use Robust Statistical Methods:** Apply methods less sensitive to outliers, such as robust regression.
