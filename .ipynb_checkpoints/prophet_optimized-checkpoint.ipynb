```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
file_path = "updated_dataset.csv" 
df = pd.read_csv(file_path)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df.head()
```

```python
df_prophet = df[['date', 'demand', 'holiday', 'school_day', 'min_temperature', 'max_temperature', 'rainfall']]
df_prophet = df_prophet.rename(columns={'date': 'ds', 'demand': 'y'})
```

```python
model = Prophet(
    changepoint_prior_scale=0.05,  # Controls flexibility, lower = less overfitting
    seasonality_mode='multiplicative',  # Try 'multiplicative' for seasonality
    yearly_seasonality=True,  # Enable yearly patterns
    weekly_seasonality=True,  # Enable weekly patterns
    daily_seasonality=False  # Disable daily seasonality to avoid noise
)
```

```python
model.add_regressor('holiday')
model.add_regressor('school_day')
model.add_regressor('min_temperature')
model.add_regressor('max_temperature')
model.add_regressor('rainfall')
```

```python
model.fit(df_prophet)
```

```python
future = model.make_future_dataframe(periods=30, freq='D')
```

```python
future = future.merge(df[['date', 'holiday', 'school_day', 'min_temperature', 'max_temperature', 'rainfall']], 
                      left_on='ds', right_on='date', how='left')
future.drop(columns=['date'], inplace=True)
```

```python
future.fillna(method='ffill', inplace=True)
```

```python
forecast = model.predict(future)
```

```python
df_results = df_prophet.merge(forecast[['ds', 'yhat']], on='ds', how='inner')
```

```python
y_true = df_results['y']
y_pred = df_results['yhat']
```

```python
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
print(f"Optimized MAE: {mae}")
print(f"Optimized RMSE: {rmse}")
print(f"Optimized MAPE: {mape:.2f}%")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_results['ds'], df_results['y'], label="Actual", marker='o')
plt.plot(df_results['ds'], df_results['yhat'], label="Predicted", linestyle='dashed')
plt.xlabel("Date")
plt.ylabel("Demand")
plt.title("Optimized Prophet Forecast vs Actual")
plt.legend()
plt.show()
```

```python

from prophet.diagnostics import cross_validation, performance_metrics

# Step 1: Try Different Hyperparameters
optimized_model = Prophet(
    changepoint_prior_scale=0.03,  # Slightly stricter to prevent overfitting
    seasonality_mode='additive',  # Test additive seasonality instead of multiplicative
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False
)

# Add same regressors
optimized_model.add_regressor('holiday')
optimized_model.add_regressor('school_day')
optimized_model.add_regressor('min_temperature')
optimized_model.add_regressor('max_temperature')
optimized_model.add_regressor('rainfall')

# Step 2: Train Model with Optimized Parameters
optimized_model.fit(df_prophet)

# Step 3: Cross-validation (Rolling Window)
df_cv = cross_validation(optimized_model, initial='365 days', period='30 days', horizon='30 days')
df_p = performance_metrics(df_cv)

print("Optimized Model Performance:")
print(df_p[['horizon', 'mae', 'rmse']])

# Step 4: Generate Forecast with Optimized Model
future_opt = optimized_model.make_future_dataframe(periods=15, freq='D')  # Reduced forecast horizon
future_opt = future_opt.merge(df[['date', 'holiday', 'school_day', 'min_temperature', 'max_temperature', 'rainfall']],
                              left_on='ds', right_on='date', how='left').drop(columns=['date'])
forecast_opt = optimized_model.predict(future_opt)

# Step 5: Evaluate Performance
df_eval_opt = df_prophet[['ds', 'y']].merge(forecast_opt[['ds', 'yhat']], on='ds', how='left')
mae_opt = mean_absolute_error(df_eval_opt['y'], df_eval_opt['yhat'])
rmse_opt = np.sqrt(mean_squared_error(df_eval_opt['y'], df_eval_opt['yhat']))

print(f"Optimized MAE: {mae_opt:.2f}")
print(f"Optimized RMSE: {rmse_opt:.2f}")

# Step 6: Plot Optimized Forecast
plt.figure(figsize=(12, 6))
plt.plot(df_eval_opt['ds'], df_eval_opt['y'], label='Actual Demand', marker='o', linestyle='-')
plt.plot(df_eval_opt['ds'], df_eval_opt['yhat'], label='Optimized Predicted Demand', marker='x', linestyle='--', alpha=0.7)
plt.xlabel("Date")
plt.ylabel("Electricity Demand")
plt.title("Optimized Prophet Forecast: Actual vs. Predicted Demand")
plt.legend()
plt.grid(True)
plt.show()
```
